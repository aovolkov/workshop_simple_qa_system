## 2. Деплой ячейки

```python
# формат ввода
input_data = {'text': 'Q: Какой-то вопрос?'}
```

```python
#deploy
#!g1.1
from transformers import GPT2Tokenizer, GPT2LMHeadModel
def generate_text(input_data):
    path_to_model = '/models/gpt3_small/checkpoint-3000'
    tokenizer = GPT2Tokenizer.from_pretrained(path_to_model)
    model = GPT2LMHeadModel.from_pretrained(path_to_model)
    model.cuda()
    inpt = tokenizer.encode(input_data['text'], return_tensors="pt")
    out = model.generate(inpt.cuda(),
                     max_length=50, 
                     repetition_penalty=5.0, 
                     do_sample=True, 
                     top_k=5, 
                     top_p=0.95, 
                     temperature=0.8,
                     stop_token='\n')
    out_decoded = tokenizer.decode(out[0])
    answer = out_decoded[len(input_data['text']):]
    return {"output_text": answer.strip('\n')}

try:
    output_data = generate_text(input_data)
except Exception as e:
    output_data = {"error": e.args[0]}
```
