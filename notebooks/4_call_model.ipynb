```python
import requests
import json

def call_node(node_id, input_params, folder_id, AIM_TOKEN):
    uri = f'https://datasphere.api.cloud.yandex.net/datasphere/v1/nodes/{node_id}:execute'
    
    data = {
        "folder_id": folder_id,
        "node_id": node_id,
        "input": input_params
    }
    headers = {
        "Host": "datasphere.api.cloud.yandex.net",
        "Accept": "application/json",
        "Content-Type": "application/json",
        'Authorization': f'Bearer {AIM_TOKEN}',
    }
    
    result = requests.post(
        uri,
        headers = headers,
        data = str(data).encode('utf-8')
    )
    
    json_result = json.loads(result.content)
    return json_result
```
