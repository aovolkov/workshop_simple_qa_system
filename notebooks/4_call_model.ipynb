# Вызов модели.

```python
import requests
import json
import os
```

```python
def call_node(node_id:str, 
              input_params:str, 
              folder_id:str, 
              token:str,
              method:str):
    "Функция для вызова ноды через IAM токен или API Key"
    
    url = f'https://datasphere.api.cloud.yandex.net/datasphere/v1/nodes/{node_id}:execute'
    
    data = {
        "folder_id": folder_id,
        "node_id": node_id,
        "input": input_params
    }
    
    if method == 'IAM':
        auth_str = f'Bearer {token}'
    elif method == 'API':
        auth_str = f'Api-Key {token}'
    else:
        raise Exception('Введите корректный метод! Используйте IAM или API.')
        
    headers = {
        "Host": "datasphere.api.cloud.yandex.net",
        "Accept": "application/json",
        "Content-Type": "application/json",
        'Authorization':auth_str,
    }
    
    result = requests.post(
        url,
        headers = headers,
        data = str(data).encode('utf-8')
    )
    
    response = json.loads(result.content)
    
    try:
        ans = response['output']['output_data']['output_text']
        return ans
    except Exception as e:
        return f'Произошла какая-то ошибка! {e.args[0]}. Status server code: {result.status_code}'
```

### Вызов через IAM.

```python
node_id = ''
input_params = {'text': 'Q: У меня не работает система умного дома? A:'}
folder_id = ''
IAM_TOKEN = ''
method = 'IAM'

call_node(node_id, input_params, folder_id, IAM_TOKEN, method)
```

### Вызов через Api-Key.

```python
node_id = ''
input_params = {'text': 'Q: У меня не работает система умного дома? A:'}
folder_id = ''
API_KEY = os.environ.get('common')
method = 'API'

call_node(node_id, input_params, folder_id, API_KEY, method)
```
