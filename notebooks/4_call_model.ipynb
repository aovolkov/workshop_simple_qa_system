# Вызов модели.

```python
import requests
import json

def call_node(node_id:str, 
              input_params:str, 
              folder_id:str, 
              IAM_TOKEN:str):
    "Функция для вызова ноды через IAM токен"
    
    url = f'https://datasphere.api.cloud.yandex.net/datasphere/v1/nodes/{node_id}:execute'
    
    data = {
        "folder_id": folder_id,
        "node_id": node_id,
        "input": input_params
    }
    headers = {
        "Host": "datasphere.api.cloud.yandex.net",
        "Accept": "application/json",
        "Content-Type": "application/json",
        'Authorization': f'Bearer {IAM_TOKEN}',
    }
    
    result = requests.post(
        url,
        headers = headers,
        data = str(data).encode('utf-8')
    )
    
    response = json.loads(result.content)
    
    try:
        ans = response['output']['output_data']['output_text']
        return ans
    except Exception as e:
        return f'Произошла какая-то ошибка! {e.args[0]}. Status server code: {result.status_code}'
```

```python
IAM_TOKEN = ''
folder_id = ''
node_id = ''
input_params = {'text': 'Q: У меня не работает система умного дома? A:'}

call_node(node_id, input_params, folder_id, IAM_TOKEN)
```
